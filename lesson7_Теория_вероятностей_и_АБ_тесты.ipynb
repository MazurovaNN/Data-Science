### Базовые понятния

**Эксперимент** - опыт, который можно повторить и который может быть закончен. *Пример* - подбрасывание монетки

**Элементарное событие** - что то произошло в мире. *Пример* - выпала решка

События бывают **достоверными**, **невозможными** и **случайными**

Достоверное событие - событие которое произойдет **всегда**. *Пример* - яблоко с дерева всегда упадет вниз

Невозможное событие - событие которое **никогда** не произойдет. *Пример* - яблоко с дерева улетит в небо

**Случайное событие** - такое событие, которое может как произойти, так и не произойти в результате эксперимента

События обозначаются заглавными латинскими буквами: $A_1, A_2, B_4$ и т.д.

**Исход** - результат эксперимента. Каждый эксперимент заканчивается исходом. *Пример* при подбрасывании монеты могут выпасть орел или решка

В простейшем случае, исходы не отличаются - они происходят или не происходят одинаково. Такие исходы называются **равновероятными**

**Вероятность наступления события A в некотором эксперименте** - Обозначается $P(A)$ и представляет собой отношение $P(A) = \frac{m}{n}$, где $n$ - число всех возможных исходов, $m$ - число исходов, *благоприятствующих* событию $A$

*Пример* - при броске монеты может выпасть или орел, или решка. Таким образом $n = 2$, орел выпадает только при одном исходе, т.е. $m = 1$, тогда $P(выпал \ орел) = \frac{m}{n} = \frac{1}{2}$

**Сумма вероятностей всех элементарных исходов эксперимента равна единице**

### Совместные и несовместные события. Алгебра событий

События называют **несовместными**, если в одном и том же испытании появление одного из событий исключает появление других событий. Такими являяются **противоположные события** (выпал орел и не выпал орел).
Противоположные события обознчаются с чертой над буквой: $\overline{A}$

**Совместные события** - события, происхождение которых не влечет за собой невозможность происхождения других. *Пример* - из колоды вытащили трефу, из колоды вытащили пику

События могут быть скомбинированны. **Запомните!**
- Союз ИЛИ означает сложение событий
- Союз И означает умножение событий

*Пример для сложения* Обозначим за $B_i$ - событие выпадения значения i на игральной кости ($i = 1 \ldots 6$),
Тогда событие "Выпало не 5" $\overline{B_5} = B_1 + B_2 + B_3 + B_4 + B_6$

*Пример для умножения*  
- $A$ - на 1-й монете выпадет орёл;
- $\overline{A}$ - на 1-й монете выпадет решка;
- $B$ - на 2-й монете выпадет орёл;
- $\overline{B}$ - на 2-й монете выпадет решка.

Тогда событие "На первой **и** второй монете выпал орел" будет равно: $AB$

**Иногда задачи при решении задач можно записать все возможные исходы и посчитать нужные**

*Пример* Чему равна вероятность при броске двух кубиков выкинуть 7?

Можно решать такую задачу просто перебрав в уме все варианты и выбрать нужные из них: но это тяжело и неудобно уже для такой задачи, что же мы будем делать с задачами сложнее?

Давайте представим все возможные исходы в виде матрицы 6x6:

```python
a = [2*i for i in range(5)]

print(a) # [0, 2, 4, 6, 8]
```

```python
# Массив в котором будут все возможные варианты
outcomes = [[i+j for j in range(1, 7)] for i in range(1, 7)]

outcomes
```

Давайте из всех исходов выберем те, значение которых равно семи и посчитаем вероятность

```python
# Счетчик
cnt_svn = 0
for i in outcomes: # i это массив
    for j in i: # j это числа
        if j == 7:
            cnt_svn += 1

# Посчитаем вероятность как отношение благоприятных исходов ко всем возможным исходам
print("Вероятность того, что сумма будет равна семи:", cnt_svn / 36)
```

## Случайные величины. Распределение случайной величины. Математическое ожидание и дисперсия случайной величины

**Случайная величина - величина, которое принимает только одно численное значение и зависит от случайных факторов** Случайную величину нельзя предугадать заранее

Случайная величина, как любая численная может быть дискретной и непрерывной.

*Дискретная случайная величина* - принимает одно из конечного (на самом деле множество может быть и бесконечным - главное, чтобы оно было счетным, подробнее тут: http://mathprofi.ru/mnozhestva.html) множества числовых возможных значений.

*Непрерывная случайная величина может принимать все значения из конечного или бесконечного интервала*

Давайте представим таблицу со значениями суммы на кубиках из предыдущего номера как последовательность случайных величин

```python
cnt = {}

for i in outcomes:
    for j in i:
        if j not in cnt.keys():
            cnt[j] = 1

        else:
            cnt[j] += 1

print(cnt)

cnt = {k : v / 36 for k, v in zip(cnt.keys(), cnt.values())}

cnt
```

Выше мы создали словарь, в котором ключи - сумма на двух кубиках и значения - кол-во таких исходов. Затем мы перевели это в вероятности - такой словарь есть **распределение вероятности случайной величины**

Построим график такого распределения создав `pd.Series` и применив к нему метод `hist`. Зададим параметр `density=True` для того чтобы получить плотность вероятности.

```python
import matplotlib.pyplot as plt
import pandas as pd

dist_cnt = pd.Series([ 2,  3,  4,  5,  6,  7,  3,  4,  5,  6,  7,  8,  4,  5,  6,  7,  8,
        9,  5,  6,  7,  8,  9, 10,  6,  7,  8,  9, 10, 11,  7,  8,  9, 10,
       11, 12])
dist_cnt.hist(density = True, bins = 11)
plt.show()
```

**Математическое ожидание** - это величина, к которой будет в среднем стремиться величина при большом количестве повторений испытаний. Мат. ожидание случайной величины $X$ обозначается как $M(X)$ или $E(X)$

Для дискретной случайной величины математическое ожидание находится по формуле: $E(X) = \sum_{i=0}^n p_ix_i$, где $x_i$ - одно из значений случайной величины, $p_i$ - вероятность выпадения этого значения

Найдем матожидание суммы при броске двух кубиков

```python
expectation = sum([k*v for k, v in zip(cnt.keys(), cnt.values())])
expectation
```

Как видно, наше мат. ожидание стремится к 7, что является наиболее вероятным значением.

Математическое ожидание показывает вокруг какого значения распределена случайная величина

**Дисперсия случайной величины**, обозначается как $D(X)$ или $Var(X)$ и находится по формуле: $$D(x) = M(X^2) - M^2(X)$$
т.е. дисперсия равна разности между матожидания матожиданием случайной величины в квадрате и квадрата матожидания случайной величины

Дисперсия показывает "разброс" величины относительно математического ожидания

## Закон больших чисел (ЗБЧ)

Закон больших чисел формулируется так: *при увеличении количества испытаний среднее арифметическое значение выборки из фиксированного распределения случайных выличин стремится к математическому ожиданию этого распределения.*

Т.е. чем больше опытов мы проводим, тем ближе среднее значение величины будет к математическому ожиданию

## Нормальное распределение. Центральная предельная теорема

Центральная предельная теорема - одна из важнейших теорем в теории вероятностей. **Она гласит:** сумма большого количества случайных величин дает *нормальное распределение*

Нормальное распределение - важнейший вид распределения, оказывается что он встречается чаще всего в реальном мире. Оно служит для описания распределения непрерывных случайных величин. Оно характеризуется двумя параметрами: средним и дисперсией: $N(\mu, \sigma^2)$


Ниже приведен график нормального распределения и демонстрация центральной предельной теоремы:

```python
import numpy as np
import pandas as pd


# Кидаем 2 кубика и смотрим на сумму
def roll_dice(n):
    a = [i for i in range(1, 7)]
    res = [np.random.choice(a) + np.random.choice(a) for i in range(n)]
    pd.Series(res).hist(bins = 11)
    plt.title(f"Распределение при {i} испытаниях")
    plt.show()
    return
```

```python
for i in range(10, 1000, 100):
    roll_dice(i)
```

## Гипотезы и p-value

Часто нам предстоит подтвердить или опровергнуть какую-либо гипотезу: отличаются ли пути покупки на нашем сайте у мужчин и женщин? Какая кнопка лучше, синяя или красная (отличается ли у них конверсионность)? и т.д.

Все начинается с формирования *нулевой гипотезы $H_0$*, например "средняя конверсия синей кнопки равна А", где А - какое то число. Однако в статистике гипотезы формулируются противоположно, так получается *альтернативная гипотеза* - "средняя конверсия синей кнопки не равна А"

Если гипотеза учитывает отклонение в обе стороны относительно наблюдаемого значения - она называется **двухсторонней** иначе - **односторонней**

Чаще всего гипотезы будут принимать следующий вид:
- среднее значение выборки равно(больше или меньше) А
- средние значения двух выборок равны между собой (одно больше другого)


**Уровень значимости или p-value** - вероятность того, что случайное наблюдение из нашего распределение будет иметь такое же или более экстремальное среднее значение по сравнению с наблюдаемым ранее, при условии что нулевая гипотеза верна


Если количество наблюдений не слишком большое (20-30), то распределение будет похоже на нормальное, но немного шире - это и есть **t-распределение Стьюдента**

Чем больше наблюдений (степеней свободы), тем больше распределение Стьюдента похоже на нормальное


Пусть нулевая гипотеза звучит так: "При бросках двух кубиков среднее значение выпавшей суммы равно 8.5" - т.е. мы предполагаем, что среднее значение выборки равно 8.5

Для подтверждения или опровержения такой гипотезы подходит статистический **t-тест** (t от t-распределения Стьюдента)

Этот и многие другие статистические тесты есть в библиотеке `scipy` в модуле `stats`

Сейчас нас интересует функция `ttest_1samp` - t-тест для одной выборки

```python
from scipy import stats as st


# Задаем среднее значение из нулевой гипотезы
null_h = 8.5

# Задаем уровень значимости
alpha_val = .05

# Массив со всеми возможными значениями эксперимента
experiments = [ 2,  3,  4,  5,  6,  7,  3,  4,  5,  6,  7,  8,  4,  5,  6,  7,  8,
        9,  5,  6,  7,  8,  9, 10,  6,  7,  8,  9, 10, 11,  7,  8,  9, 10,
       11, 12]

# Получаем результаты t-теста
res = st.ttest_1samp(experiments, null_h)

print(res.pvalue)

# Отвергаем нулевую гипотезу если p-value < alpha, иначе принимаем
if res.pvalue < alpha_val:
    print("Отвергаем нулевую гипотезу")

else:
    print("Не можем отвергнуть нулевую гипотезу")
```

Чаще всего требуется проверить равенство средних (или других статистических параметров) в двух независимых или зависимых выборках.

Для того, чтобы понять есть ли статистически значимая разница между двумя независимыми выборками также можно пользоваться t-тестом. Для такого случая используют функцию `ttest_ind`.

Для зависимых выборок (например средняя выручка до и после появления новых товаров) используется функция `ttest_rel`

```python
new_experiments = [i for i in range(36)]

res = st.ttest_ind(experiments, new_experiments, equal_var=False)

# H0 - средние двух выборок равны между собой
alpha = .05

print(res.pvalue)

# Отвергаем нулевую гипотезу если p-value < alpha, иначе принимаем
if res.pvalue < alpha_val:
    print("Отвергаем нулевую гипотезу")

else:
    print("Не можем отвергнуть нулевую гипотезу")
```

## Доверительные интервалы и Bootstrap

Стандартная ошибка среднего (Standard error of the mean) вычислятся по формуле: $$\frac{\sigma}{\sqrt{n}}$$

```python
import numpy as np
import scipy.stats


def confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    # среднее значение и стандартная ошибка среднего
    m, se = np.mean(a), scipy.stats.sem(a)

    # функция квантилей
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
def bootstrap(a, n, s):
    res = []
    for i in range(n):
        b = np.random.choice(a, s)
        res.append(np.mean(b))

    return res
```

```python
a = [i for i in range(100)]

k = bootstrap(a, 1000, 30)
```
